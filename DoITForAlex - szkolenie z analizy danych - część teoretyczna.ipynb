## 2. Pandas

````python
import pandas as pd
````

Podstawowa struktura danych w pandas to **Series**, którą można tworzyć z list, słowników, array'ów. Series odpowiada jednej kolumnie w tabeli danych.

````python
lista_1 = [1, 2, 3]
pd.Series(data=lista_1)
````

````python
lista_2 = ['a', 'b', 'c']
pd.Series(data=lista_2)
````

````python
lista_3 = [1, 2, 'c', 4]
pd.Series(data=lista_3)
````

````python
pd.Series(data=lista_2, index=lista_1)
````

````python
dict_1 = {1:'a', 2:'b', 3:'c'}
pd.Series(data=dict_1)
````

````python
ser1 = pd.Series([0,5,1,4],index = ['rower', 'kajaki','kite', 'basen']) 
print(ser1)
ser2 = pd.Series([1,0,5,2],index = ['rower', 'kajaki','biegi', 'basen'])  
print(ser2)

# odwołanie po indexie
ser1['rower']
````

### Brak danych

````python
# brak danych - wartości NaN

df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
df
````

````python
# usuwa wiersze jeżeli liczba wartości nieznanych jest wyższa niż liczba odcięcia

df.dropna(thresh=2)
````

### Group by

````python
data = {'Firma':['McDonalds', 'McDonalds','KFC','KFC','Starbucks','Starbucks'],
       'Osoba':['Ania','Kasia','Piotr','Mateusz','Basia','Szymon'],
       'Sprzedaż':[400,250,222,333,200,120]}
df = pd.DataFrame(data)
df
````

````python
by_firma = df.groupby('Firma')
````

````python
by_firma
````

````python
df.groupby('Firma').mean()
````
