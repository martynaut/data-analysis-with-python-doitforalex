````python
df_left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df_right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
df_left
````

````python
df_right
````

````python
# merge
pd.merge(df_left, df_right, how='inner',on='key')
````

````python
df_left.join(df_right, rsuffix='right')
````
